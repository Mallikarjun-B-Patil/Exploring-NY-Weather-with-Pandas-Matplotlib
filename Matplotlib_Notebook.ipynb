# **DATA VISUALIZATION - MATPLOTLIB**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Matplotlib Plotting

```python
# Simple plot

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

```python
# Plotting without line

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y, "o")
plt.show()
```

```python
# Multiple line plot

x = np.array([0,2,6,9])
y = np.array([0,7,3,8])

plt.plot(x, y)
plt.show()
```

```python
# Default x-points
# If x value is not given, it considers [0,1,2,3....] as x-points

y = np.array([0,7,4,9,1,6,0])

plt.plot(y)
plt.show()
```

# Markers

```python
# Use marker to mark the points in plot

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "o")
plt.show()
```

```python
# Different marking Points

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "d")
plt.show()
```

```python
# Marker size (ms)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "o", ms= 18)
plt.show()
```

```python
# Marker Edge Colour (mec)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "o", ms= 18, mec= "r")
plt.show()
```

```python
# Marker Face Colour (mfc)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "o", ms= 18, mfc= "y")
plt.show()
```

```python
# Using both mec and mfc

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, marker= "o", ms= 18, mec= "m", mfc= "m")
plt.show()
```

# Lines

```python
# Line Styles (ls or linestyle)
# Use either name ("dashed") or symbol ("--")

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, ls= "dashed")
plt.show()
```

```python
# Line Colour (color or c)
# Use either name (color) or symbol (c)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, c= "c")
plt.show()
```

```python
# Line Width (linewidth or lw)
# Use either name (linewidth) or symbol (lw)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, lw= 10)
plt.show()
```

```python
# Multiple Lines

x = np.array([0,1,2,3,4])
y1 = np.array([0,7,4,9,1])
y2 = np.array([9,1,3,2,7])

plt.plot(x, y1, y2)
plt.show()
```

```python
# Multiple Lines of more X values

x1 = np.array([0,1,2,3,4])
x2 = np.array([4,3,2,1,0])
y1 = np.array([0,7,4,9,1])
y2 = np.array([9,1,3,2,7])

plt.plot(x1, y1, x2, y2)
plt.show()
```

# Combining Lines, Pointers and Colors

```python
# Different marking Lines - Dashed line with red colour
# "o" - Marking Point. "-" - Marking Line. "r" -Colour (r-Red)

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, "o--r")
plt.show()
```

```python
# Dotted line with blue colour

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, "o:b")
plt.show()
```

```python
# Dashed Dotted line with green colour

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,7,4,9,1,6,0])

plt.plot(x, y, "o-.g")
plt.show()
```

# Labels and Title

```python
# Label the graph with "xlabel" (X-Axis) and "ylabel" (Y-Axis)
# Write name of the graph with "title"

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.plot(x, y)

plt.title("Products sold on days of Week")
plt.xlabel("Days of Week")
plt.ylabel("Products Sold")

plt.show()
```

```python
# "size=" - Size of the Label and Title
# "loc=" - Location of the Label Title ("left", "right", "center"=default)

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.plot(x, y)

plt.title("Products sold on days of Week", size= 20, loc="left")
plt.xlabel("Days of Week", size= 15)
plt.ylabel("Products Sold", size= 15)

plt.show()
```

# Adding Grid Lines

```python
# "grid()" - Adding Grid lines to the plot

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.plot(x, y)

plt.title("Products sold on days of Week")
plt.xlabel("Days of Week")
plt.ylabel("Products Sold")

plt.grid()

plt.show()
```

```python
# Adding Grid lines to the perticular Axis

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.plot(x, y)

plt.title("Products sold on days of Week")
plt.xlabel("Days of Week")
plt.ylabel("Products Sold")

plt.grid(axis= "x")

plt.show()
```

```python
# Adding Grid lines with properties

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.plot(x, y)

plt.title("Products sold on days of Week")
plt.xlabel("Days of Week")
plt.ylabel("Products Sold")

plt.grid(c= "g", ls= "--", lw= "0.5")

plt.show()
```

# Scatter Plots

```python
# Scatter plots - Draws dots

x = np.array([1,2,3,4,5,6,7])
y = np.array([100,750,480,910,290,650,80])

plt.scatter(x, y)
plt.show()
```

```python
# Comparing two scatter plots

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x, y)

plt.show()
```

```python
# Colour variation along with Colorbar

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=color, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
# Dot Sizes

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, s=sizes, alpha=0.8)

plt.show()
```

# Bar Plots

```python
# Simple Bar Plot

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 5, 10])

plt.bar(x,y)
plt.show()
```

```python
# Horizontal Bars - (barh)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 5, 10])

plt.barh(x,y)
plt.show()
```

```python
# Vertical bar (width)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 5, 10])

plt.bar(x, y, color="c", width= 0.5)
plt.show()
```

```python
# Horizontal bar (height)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 5, 10])

plt.barh(x, y, color="c", height= 0.5)
plt.show()
```

# Pie Charts

```python
# Simple Pie Chart

arr = np.array([20, 25, 15, 30, 10])

plt.pie(arr)
plt.show() 
```

```python
# Pie Chart with Labels

arr = np.array([20, 25, 15, 30, 10])
fruits = ["Apples", "Bananas", "Cherries", "Oranges", "Mangos"]

plt.pie(arr, labels= fruits)
plt.show() 
```

```python
# Start angle

arr = np.array([20, 25, 15, 30, 10])
fruits = ["Apples", "Bananas", "Cherries", "Oranges", "Mangos"]

plt.pie(arr, labels= fruits, startangle=90)
plt.show() 
```

```python
# Explode with Shadow

arr = np.array([20, 25, 15, 30, 10])
fruits = ["Apples", "Bananas", "Cherries", "Oranges", "Mangos"]
space = [0.2,0,0,0,0]

plt.pie(arr, labels= fruits, explode= space, shadow= True)
plt.show() 
```

```python
# Complete Pie Chart

arr = np.array([20, 25, 15, 30, 10])
fruits = ["Apples", "Bananas", "Cherries", "Oranges", "Mangos"]

plt.pie(arr, labels= fruits, explode= space, shadow= True)
plt.title("Pie Chart")
plt.legend(title= "Fruits List")
plt.show() 
```

```python
arr = np.array([20, 25, 15, 30, 10])
fruits = ["Apples", "Bananas", "Cherries", "Oranges", "Mangos"]

plt.pie(arr, labels= fruits)
plt.title("Pie Chart")
plt.legend()
plt.show() 
```
