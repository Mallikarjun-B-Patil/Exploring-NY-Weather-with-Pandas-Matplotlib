```python
# Filling all null values
# mean()

avg_temp = df["Temperature"].mean()
df["Temperature"].fillna(avg_temp)
```

```python
# Replacing the filled values in Dataset

avg_temp = df["Temperature"].mean()
df["Temperature"].fillna(avg_temp, inplace=True)
df
```

```python
# median()

med_dew = df["DewPoint"].median()
df["DewPoint"].fillna(med_dew, inplace=True)
df
```

```python
# mode()[0]

vis_mod = df["VisibilityMiles"].mode()[0]
df["VisibilityMiles"].fillna(vis_mod, inplace=True)
df
```

```python
# Forward fill - ffill()

ffill_sea = df["Sea Level PressureIn"].ffill()
df["Sea Level PressureIn"].fillna(ffill_sea, inplace=True)
df
```

```python
# Backward fill - bfill()

bfill_hum = df["Humidity"].bfill()
df["Humidity"].fillna(bfill_hum, inplace=True)
df
```

```python
# Fill with constant value

df["WindSpeedMPH"].fillna(6.5, inplace=True)
df
```

```python
# Replace string "NaN" with actual NaN, if present

df["Event"].replace("NaN", pd.NA, inplace=True)
df["Event"].fillna("Drizzle", inplace=True)
df
```

```python
# Mean of numeric columns

df.mean(numeric_only=True)
```

```python
# Filling all null at a time

df.fillna({"Temperature":36.5,"DewPoint":22.9,"Humidity":50.8,"Sea Level PressureIn":30.4,"VisibilityMiles":9.09,
           "WindSpeedMPH":6.54},inplace=True)
df
```

# Combining the files

```python
# Reading the students file

students = pd.read_csv("students.csv")
students
```

```python
# Reading the students file

professors = pd.read_csv("professors.csv")
professors
```

```python
# Saving the csv file with header and not to include the index column

students.to_csv("students.csv",index=False,header=True)
students
```

```python
# Saving the csv file with header and not to include the index column

professors.to_csv("professors.csv",index=False,header=True)
professors
```

```python
# Concatenation of both DataFrames

pd.concat([students, professors], ignore_index = True)
```

```python
# Matching and Merging both DataFrames

pd.merge(students, professors[['id_number', 'Prof_name', 'Department']])
```

```python
# Left Join of both DataFrames
# Ensures that no rows are lost from the left DataFrame

pd.merge(students, professors, how = 'left')
```

```python
# Right Join of both DataFrames
# Ensures that no rows are lost from the right DataFrame

pd.merge(students, professors, how = 'right')
```

```python
# Outer Join of both DataFrames
# Combines all rows from both DataFrames, Includes rows that don't have matching values

pd.merge(students, professors, how = 'outer')
```

```python
# Inner Join of both DataFrames
# Includes only the rows where there is a match, Discards rows that do not have a match

pd.merge(students, professors, how = 'inner')
```

```python
# Combining two DataFrames by filling missing values

students.combine_first(professors)
```
